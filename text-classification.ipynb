```python
import numpy as np

def generate_predictions(sequence, categories):
    result = pipe(sequence, candidate_labels=categories)
    max_index = np.argmax(result["scores"])
    max_label = result["labels"][max_index]
    return max_label
```

```python
from tqdm import tqdm
import numpy as np

# Make sure fiction_categories and generate_predictions are defined
fiction_categories = ["Fiction", "Nonfiction"]

def generate_predictions(sequence, categories):
    result = pipe(sequence, candidate_labels=categories)
    max_index = np.argmax(result["scores"])
    return result["labels"][max_index]

# Initialize lists
actual_cats = []
predicted_cats = []

# Get 300 Fiction samples
fiction_texts = books.loc[books["simple_categories"] == "Fiction", "description"].dropna().reset_index(drop=True)

for i in tqdm(range(min(300, len(fiction_texts)))):
    sequence = fiction_texts[i]
    predicted_cats.append(generate_predictions(sequence, fiction_categories))
    actual_cats.append("Fiction")

# Get 300 Nonfiction samples
nonfiction_texts = books.loc[books["simple_categories"] == "Nonfiction", "description"].dropna().reset_index(drop=True)

for i in tqdm(range(min(300, len(nonfiction_texts)))):
    sequence = nonfiction_texts[i]
    predicted_cats.append(generate_predictions(sequence, fiction_categories))
    actual_cats.append("Nonfiction")
```
